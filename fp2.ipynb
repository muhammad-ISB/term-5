```python
import pandas as pd
import json

def read_json_file(file_path):
    dataframes = []

    with open(file_path, 'r') as file:
        for line in file:
            json_obj = json.loads(line.strip())

            # Check if 'data' and 'name' keys exist in the JSON object
            if 'data' in json_obj and 'name' in json_obj:
                df = pd.DataFrame(json_obj['data'], columns=['Date', json_obj['name']])
                df['Date'] = pd.to_datetime(df['Date'], format='%Y%m', errors='coerce')  # Handle different date formats
                df = df.dropna().set_index('Date')
                dataframes.append(df)
            else:
                # Handle cases where 'data' or 'name' keys are missing
                # For example, you can print a warning or pass
                print(f"Warning: Skipping a line due to missing 'data' or 'name' key.")

    combined_df = pd.concat(dataframes, axis=1)
    return combined_df

# Usage
file_path = '/content/drive/MyDrive/fp2/TOTAL Energy.txt'  # Replace with your file path
combined_df = read_json_file(file_path)
```

```python
mixed_type_columns = []
missing_values_info = {}

for column in combined_df.columns:
    if combined_df[column].apply(type).nunique() > 1:  # Checking for mixed types
        mixed_type_columns.append(column)
    missing_values_info[column] = combined_df[column].isna().sum()
```

```python
# Step 1: Handling Mixed Data Types
energy_data_numeric = combined_df.apply(pd.to_numeric, errors='coerce')

# Step 2: Dealing with Missing Values
# Forward fill to propagate last valid observation forward
energy_data_numeric_ffill = energy_data_numeric.fillna(method='ffill')

# Then fill remaining missing values with the mean of each column
energy_data_cleaned = energy_data_numeric_ffill.fillna(energy_data_numeric_ffill.mean())

# Checking the first few rows of the cleaned DataFrame
energy_data_cleaned
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Selecting the column for forecasting
column_to_forecast = 'Electricity Net Generation Total (including from sources not shown), All Sectors, Monthly'
series = energy_data_cleaned[column_to_forecast].dropna()
```

```python
# SARIMA Model parameters (p, d, q) and seasonal parameters (P, D, Q, S)
p, d, q = 1, 1, 1
P, D, Q, S = 1, 1, 1, 12  # Assuming a yearly seasonality

# Fitting the SARIMA model
model = SARIMAX(series, order=(p, d, q), seasonal_order=(P, D, Q, S), enforce_stationarity=False, enforce_invertibility=False)
sarima_model = model.fit(disp=False)

# Forecasting the next 12 months
forecast = sarima_model.get_forecast(steps=12)
forecast_index = pd.date_range(series.index[-1], periods=13, closed='right', freq='M')
forecast_series = pd.Series(forecast.predicted_mean, index=forecast_index)
```

```python
# Plotting the historical data and forecast
plt.figure(figsize=(12, 6))
plt.plot(series, label='Historical Data')
plt.plot(forecast_series, label='Forecast', color='red')
plt.title(f'Forecast for {column_to_forecast}')
plt.xlabel('Date')
plt.ylabel(column_to_forecast)
plt.legend()
plt.grid(True)
plt.show()
```

```python
next_column_to_forecast = 'Electricity End Use, Total, Monthly'
series_next = energy_data_cleaned[next_column_to_forecast].dropna()
```

```python
# Fitting the SARIMA model to the next column
model_next = SARIMAX(series_next, order=(p, d, q), seasonal_order=(P, D, Q, S), enforce_stationarity=False, enforce_invertibility=False)
sarima_model_next = model_next.fit(disp=False)
```

```python
# Forecasting the next 12 months for the next column
forecast_next = sarima_model_next.get_forecast(steps=12)
forecast_index_next = pd.date_range(series_next.index[-1], periods=13, closed='right', freq='M')
forecast_series_next = pd.Series(forecast_next.predicted_mean, index=forecast_index_next)
```

```python
# Plotting the historical data and forecast for the next column
plt.figure(figsize=(12, 6))
plt.plot(series_next, label='Historical Data')
plt.plot(forecast_series_next, label='Forecast', color='red')
plt.title(f'Forecast for {next_column_to_forecast}')
plt.xlabel('Date')
plt.ylabel(next_column_to_forecast)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Remaining selected columns for forecasting
remaining_columns = [
    'Electricity Net Generation Total (including from sources not shown), Electric Power Sector, Monthly',
    'Electricity Sales to Ultimate Customers, Total, Monthly',
    'Electricity Net Generation From Natural Gas, All Sectors, Monthly'
]

# Dictionary to hold forecasts for remaining columns
forecasts = {}

# Iterating over the remaining columns and generating forecasts
for column in remaining_columns:
    series_remaining = energy_data_cleaned[column].dropna()
    model_remaining = SARIMAX(series_remaining, order=(p, d, q), seasonal_order=(P, D, Q, S), enforce_stationarity=False, enforce_invertibility=False)
    sarima_model_remaining = model_remaining.fit(disp=False)

    # Forecasting the next 12 months for each remaining column
    forecast_remaining = sarima_model_remaining.get_forecast(steps=12)
    forecast_index_remaining = pd.date_range(series_remaining.index[-1], periods=13, closed='right', freq='M')
    forecast_series_remaining = pd.Series(forecast_remaining.predicted_mean, index=forecast_index_remaining)

    forecasts[column] = forecast_series_remaining

# Displaying the forecast plots for each of the remaining columns
plt.figure(figsize=(15, 10))

for i, (column, forecast_series) in enumerate(forecasts.items(), 1):
    plt.subplot(len(forecasts), 1, i)
    plt.plot(energy_data_cleaned[column].dropna(), label='Historical Data')
    plt.plot(forecast_series, label='Forecast', color='red')
    plt.title(f'Forecast for {column}')
    plt.xlabel('Date')
    plt.ylabel(column)
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
